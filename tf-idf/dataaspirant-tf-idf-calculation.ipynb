# **TF-IDF Calculation**



### Import Requried **Python** Packages

```python
import pandas as pd
import math
```

#### Let's say we have **3 documents** having 3 strings each.

```python
doc1 = ["Java is to JavaScript what car is to Carpet", 
        "I learned the fundamentals for programming, which is just what I needed as a first step for my career change", 
        "Learning to code is useful no matter what your career ambitions are"]

doc2 = ["First, solve the problem. Then, write the code", 
        "I came in with near zero programming knowledge and halfway in, I am quite confident of what I can achieve",
        "When you throw yourself out there great things happen"]
        
doc3 = ["Experience is the name everyone gives to their mistakes", 
        "Coding is the language of the future, and every girl should learn it", 
        "if you are good at it, it can help you achieve your dreams"]
```

### Create **tokens** for each document.

Convert to **lower string** and replace commas

```python
doc1_tokens = sum([doc.lower().replace(',', '').replace('.', '').split() for doc in doc1], [])
doc2_tokens = sum([doc.lower().replace(',', '').replace('.', '').split() for doc in doc2], [])
doc3_tokens = sum([doc.lower().replace(',', '').replace('.', '').split() for doc in doc3], [])

print(doc1_tokens)
print('\n', doc2_tokens)
print('\n', doc3_tokens)
```

#### Remove stop words

```python
stopwords = ['a', 'the', 'i', 'me', 'myself', 'our', 'ourself', 'you', 'is', 'to', 'then', 'what', 'their', 'are', 'your', 
             'for', 'my', 'as', 'which', 'just', 'can', 'and', 'in', 'of', 'am', 'when', 'there', 'at', 'it', 'if']

doc1_tokens = set(doc1_tokens) - set(stopwords)
doc2_tokens = set(doc2_tokens) - set(stopwords)
doc3_tokens = set(doc3_tokens) - set(stopwords)

print(doc1_tokens)
print(doc2_tokens)
print(doc3_tokens)
```

#### Find the **unique** set of tokens.

```python
unique_tokens = set(doc1_tokens).union(set(doc2_tokens)).union(set(doc3_tokens))

print(unique_tokens)
```

#### Maintain a dict to keep track of count of the unique words in the individual docs

```python
count_doc1 = dict.fromkeys(unique_tokens, 0)
for token in doc1_tokens:
    count_doc1[token] += 1
    
count_doc2 = dict.fromkeys(unique_tokens, 0)
for token in doc2_tokens:
    count_doc2[token] += 1
    
count_doc3 = dict.fromkeys(unique_tokens, 0)
for token in doc3_tokens:
    count_doc3[token] += 1
    
print(count_doc1)
print('\n', count_doc2)
print('\n', count_doc3)
```

### Calculate term frequency **(TF)**: 

Word frequency divided by number of words in the doc

```python
def calculate_tf(count_doc, doc_tokens):
    tf = dict()
    for token, count in count_doc.items():
        tf[token] = count / float(len(doc_tokens))
    return tf
    
tf1 = calculate_tf(count_doc=count_doc1, doc_tokens=doc1_tokens)
tf2 = calculate_tf(count_doc=count_doc2, doc_tokens=doc2_tokens)
tf3 = calculate_tf(count_doc=count_doc3, doc_tokens=doc3_tokens)

print(tf1)
print('\n', tf2)
print('\n', tf3)
```

### Calculate inverse document frequency **IDF**

Log of number of documents divided by the number of documents containing a specific word. 

```python
def calculate_idf(doc_counts):
    idf = dict.fromkeys(doc_counts[0].keys(), 0)
    for doc in doc_counts: 
        for token, count in doc.items():
            if count!=0:
                idf[token] += 1
                
    for token, count in idf.items():
        idf[token] = math.log(len(doc_counts) / float(count))
       
    return idf

idf = calculate_idf([count_doc1, count_doc2, count_doc3])
print(idf)
```

### Calculate TF-IDF

Multiply TF and IDF

```python
def calculate_tfidf(tf, idf):
    tfidf = dict()
    for token, count in tf.items():
        tfidf[token] = count * idf[token]
    return tfidf
```

```python
tfidf1 = calculate_tfidf(tf1, idf)
tfidf2 = calculate_tfidf(tf2, idf)
tfidf3 = calculate_tfidf(tf3, idf)
```

#### Create a dataframe for all the calculated values

```python
tfidf_df = pd.DataFrame([tfidf1, tfidf2, tfidf3])
tfidf_df
```
